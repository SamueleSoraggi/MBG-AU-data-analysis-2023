# Small example in R

## Generate a small dataset

```r
set.seed(0)
x <- seq(0, 10, length.out = 20)
y <- 2 * x + 1 + rnorm(20)
```

## Calculate coefficients for the regression line

```r
regression <- lm(y ~ x) #regression
slope <- regression$coefficients[2]
intercept <- regression$coefficients[1]
```

## Calculate the regression line

```r
regression_line <- slope * x + intercept
```

## Plot the dataset and regression line

```r
plot(x, y, main = "Scatterplot with Regression Line", xlab = "X", ylab = "Y")
abline(regression, col = "red")
legend("topleft", legend = "Regression Line", col = "red", lty = 1)
```
