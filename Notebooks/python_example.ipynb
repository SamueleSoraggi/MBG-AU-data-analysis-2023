# Small example in python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Generate a small dataset

```python
np.random.seed(0)
x = np.linspace(0, 10, 20)
y = 2 * x + 1 + np.random.normal(0, 1, 20)
```

## Calculate coefficients for the regression line

```python
X = np.column_stack([x, np.ones_like(x)])  # Add a column of ones for the intercept
coefficients, residuals, _, _ = np.linalg.lstsq(X, y, rcond=None)
slope, intercept = coefficients
```

## Calculate the regression line

```python
regression_line = slope * x + intercept
```

## Plot the dataset and regression line

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Data')
plt.plot(x, regression_line, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot with Regression Line')
plt.legend()
plt.grid(True)
plt.show()
```
